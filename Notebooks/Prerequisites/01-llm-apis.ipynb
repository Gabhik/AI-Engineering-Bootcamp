```python
# OpenAI API Example
import openai
import os

try:
    client = openai.OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
    
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": "can you print 10 random words?"}
        ]
    )
    
    print("OpenAI Response:")
    print(response.choices[0].message.content)
    
except openai.RateLimitError as e:
    print(f"‚ö†Ô∏è OpenAI Quota Error: {e}")
    print("üí° Your OpenAI API key has exceeded its quota or needs billing setup.")
    print("   Check: https://platform.openai.com/account/billing")
except Exception as e:
    print(f"‚ùå OpenAI Error: {e}")
```

```python
# Groq API Example (Alternative - Fast & Free tier available)
from groq import Groq
import os

try:
    client = Groq(api_key=os.getenv("GROQ_API_KEY"))
    
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",  # Fast and free tier available
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": "can you print 10 random words?"}
        ]
    )
    
    print("Groq Response:")
    print(response.choices[0].message.content)
    
except Exception as e:
    print(f"‚ùå Groq Error: {e}")
```

```python
# Google Gemini API Example (Alternative)

from google import genai

gemini_client = genai.Client(api_key=os.getenv("GOOGLE_API_KEY"))

response = gemini_client.models.generate_content(
    model="gemini-2.5-flash",
    contents=["You are a helpful assistant",
              "can you print 15 random words?"]
)

print(response.text)
```
